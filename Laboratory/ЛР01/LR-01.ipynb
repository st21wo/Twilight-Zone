```python
# функция шифрования по алгоритму цезаря
def tsesar():
    # для работы необходим алфавит, объявляем его
    letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZABCDEFGHIJKLMNOPQRSTUVWXYZАБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
    # задаем шаг шифровки
    step = 5
    # строка ввода текста
    text = input("Цезарь - шифрование :)")
    # переменная для записи результата
    result = ''
    # процесс шифрования 
    for i in text:
        ind = letters.find(i) # здесь вычисляются места символов в списке
        newind = ind + step   # здесь символы сдвигаются на заданный в переменной step шаг
        if i in letters:
            result += letters[newind]  # задаем значение в переменную для записи результата 
        else:
            result += i
    print(result)
```

```python
#дешифровка
def tsesar_deshifr():
    letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZABCDEFGHIJKLMNOPQRSTUVWXYZАБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
    smeshenie = 5
    text = input("Цезарь - дешифровка")
    result = ''

    for i in text:
        ind = letters.find(i)
        newind = ind - smeshenie
        if i in letters:
            result += letters[newind]
        else:
            result += i
    print(result)
```

```python
# шифр атбаша заключается в том, что меняются буквы из обычного алфавита на буквы из алфавита-наоборот
def atbash():
    # задаем алфавит
    letters = [chr(x) for x in range(65, 91)]
    # алфавит-наоборот
    letters_r = [x for x in letters]
    letters_r.reverse()

    text = input("Атбаш - шифрование")
    result = ""
    # перебираются буквы из исходного текста
    for i in text:
        # перебираются индексы и значения из letters
        for j,l in enumerate(letters):
            if i == l: # если буквы i и l равны, то
                result += letters_r[j] # ставим в результат букву из реверсированного списка с индексом j
    print(result)
```

```python
# функция дешифровки 
# меняются местами списки чтобы дешифровать сообщение 
def atbash_desh():
    letters = [chr(x) for x in range(65, 91)]
    letters_r = [x for x in letters]
    letters_r.reverse()

    text = input("Атбаш - дешивровка")
    result = ""
    for i in text:
        for j, l in enumerate(letters_r):
            if i == l:
                result += letters[j]
    print(result)
```
